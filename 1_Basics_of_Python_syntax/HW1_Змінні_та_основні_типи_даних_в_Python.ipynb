**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення

```python
var1 = 15
var2 = 4

print(f"""Сума: {var1 + var2}
Різниця: {var1 - var2}
Добуток: {var1 * var2}
Ділення: {var1 / var2}
Цілочисельне ділення: {var1 // var2}
Залишок від ділення: {var1 % var2}
Піднесення до степеня: {var1 ** var2}""")
```
