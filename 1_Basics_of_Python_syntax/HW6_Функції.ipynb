**1. Функція без параметрів**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
# функція з параметром
def show_menu(dishes_list):
    print('=== МЕНЮ ===')
    for i, dish in enumerate(dishes_list):
        print(f'{i+1}. {dish}')
    print('=============')

my_dishes = ["Суп грибний", "Котлета по-київськи", "Лимонад"]
show_menu(my_dishes)

# # функція без параметрів
# def show_menu():
#     dishes = ["Суп грибний", "Котлета по-київськи", "Лимонад"]
#     print('=== МЕНЮ ===')
#     for i, dish in enumerate(dishes):
#         print(f'{i+1}. {dish}')
#     print('=============')
# show_menu()
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    print(f'Квадрат числа {n} = {n**2}')

print_square(5)
print_square(10)
print_square(7)
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
# пишемо свою функцію аналог до max
def find_max(numbers):
    max_numb = numbers[0]
    for n in numbers:
        if (n > max_numb):
            max_numb = n
    return max_numb

my_list = [15, 42, 8, 23, 56, 19]
print(f'Максимальне значення зі списку {my_list} є {find_max(my_list)}')
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade):
    if grade >=60:
        return "Здав"
    else:
        return "Не здав"

my_grade1 = 60
print(f'Оцінка {my_grade1}: {check_grade(my_grade1)}')
my_grade2 = 83
print(f'Оцінка {my_grade2}: {check_grade(my_grade2)}')
my_grade3 = 45
print(f'Оцінка {my_grade3}: {check_grade(my_grade3)}')
````
