**Про датасет:**

Tips - це датасет, який містить інформацію про рахунки та чайові в ресторані.

**Колонки датасету:**
- `total_bill` - загальна сума рахунку (долари)

- `sex` - стать відвідувача (Male/Female)

- `time` - час відвідування (Lunch/Dinner)
- `size` - розмір групи (кількість людей)

**Джерело:** Оригінальні дані зібрані в одному з американських ресторанів у 1990-х роках. Скачати і переглянути набір даних можна тут: https://www.kaggle.com/datasets/sakshisatre/tips-dataset

**4. Статистика**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній розмір рахунку на всьому наборі даних
- Максимальні чайові
- Кількість пропущених значень у всіх колонках
- Скільки відвідувачів обідали, скільки вечеряли

```python
print('Середній розмір рахунку на всьому наборі даних:', df['total_bill'].mean())
print('Максимальні чайові:', df['tip'].max())
```

```python
# Кількість пропущених значень у всіх колонках:
df.isnull().sum()
```

```python
# Тут ми знаходимо к-сть рахунків за вечері і за обіди 
lunch_count = df['time'] == 'Lunch'
dinner_count = df['time'] == 'Dinner'
print(f'Було здійснено {lunch_count.sum()} рахунків за обіди, {dinner_count.sum()} рахунків за вечері.')
display(df['time'].value_counts()) # або рахуємо к-сть рахунків для кожного значення time, так як у нас їх тільки 2: Lunch і Dinner

# Також порахуємо з урахуванням розмірів груп (size)
print(f'Обідало {df[lunch_count]['size'].sum()} відвідувачів. Вечеряло {df[dinner_count]['size'].sum()} відвідувачів.')
display(df.groupby('time')['size'].sum()) # або рахуємо к-сть відвідувачів кожного значення time, так як у нас їх тільки 2: Lunch і Dinner
```

**5. Групування**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день


```python
# Середній рахунок за статтю
df.groupby('sex')['total_bill'].mean()
```

```python
# Середні чайові за днем тижня
df.groupby('day')['tip'].mean()
```

```python
# Скільки відвідувачів у кожен день
display(df.groupby('day')['size'].sum())

# Таким чином підраховуємо рахунки, але не враховуємо розміри груп
# display(df.groupby(['day']).size())
```

**6. Створення нової колонки**

Створіть нову колонку `tip_percentage` - відсоток чайових від загальної суми рахунку.
Виведіть перші 10 рядків з новою колонкою.

```python
df['tip_percentage'] = (df['tip'] / df['total_bill'] * 100).round(2)
df.head(10)
```
