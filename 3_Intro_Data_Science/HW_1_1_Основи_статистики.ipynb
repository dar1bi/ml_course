  6.1. без використання жодних бібліотек, користуючись тільки Python.

Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
# без використання жодних бібліотек, користуючись тільки Python

# розрахунки для visits (для sum_purchases вже маємо)
mean_visits = sum(visits)/len(visits)
visits_minus_mean = [(x - mean_visits)**2 for x in visits]
std_visits = (sum(visits_minus_mean)/(len(visits)-1))**0.5
# знаходимо коваріацію
purchases_visits = [ (sum_purchases[i] - mean_purchases)*(visits[i] - mean_visits) for i in range(n_purchases)]
cov_purchases_visits = sum(purchases_visits)/(n_purchases-1)
# знаходимо коефіцієнт кореляції Пірсона
corr_purchases_visits = cov_purchases_visits / (std_purchases * std_visits)
# або відразу наступною формулою:
# corr_purchases_visits = sum(purchases_visits) / ((sum(purchases_minus_mean))**0.5 * (sum(visits_minus_mean))**0.5)


# використовуючи відповідний метод бібліотеки numpy
corr_purchases_visits_np = np.corrcoef(sum_purchases, visits)

print(corr_purchases_visits, corr_purchases_visits_np[0, 1])
# Бачимо, що результати співпадають
```

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

***Коефіцієнт -0.3 вказує на негативну, але слабку лінійну кореляцію. Дивлячись на графік розсіювання, бачимо на ньому викиди. Отже є сенс розрахувати окремо коефіцієнт кореляції Пірсона не враховуючи ці викиди.***

Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
purchases_arr = np.array(sum_purchases)
visits_arr = np.array(visits)

threshold = mean_purchases + 2 * std_purchases
mask = purchases_arr <= threshold

purchases_new = purchases_arr[mask]
visits_new = visits_arr[mask]

new_corr = np.corrcoef(purchases_new, visits_new)[0, 1]
print(new_corr)
```
