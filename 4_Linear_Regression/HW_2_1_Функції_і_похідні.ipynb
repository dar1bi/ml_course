1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')

f = 10*x - 5
df = sp.diff(f, x)
print("Функція:", f)
print("Похідна:", df)
```

```python
f_num = sp.lambdify(x, f, 'numpy')
df_num = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_num(x_vals), label=f'f(x) = {f}', color='blue')
plt.plot(x_vals, np.full_like(x_vals, df_num(x_vals)), label=f"f'(x) = {df}", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Оцінка має лінійну залежність від витрачених годин на навчання, а саме $ f(x) = 10x - 5 $ , де $ f(x) $ - оцінка, $ x $ - години навчання.  
Для прикладу, витративши 2 години навчання $ (x = 2) $, оцінка буде $ f(2) = 10*2 - 5 = 15 $   
А витративши 5 годин $ (x = 5) $, оцінка буде $ f(5) = 10*5 - 5 = 45 $

Значення похідної $ f'(x) = 10 $, показує, що в будь-якій точці вона не змінюється, це означає, що при збільшенні кількості годин навчання на 1 значення оцінки зростає на 10 балів незалежно від точки $ x $.


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)
print("Функція:", f)
print("Похідна:", df)
```

```python
f_num = sp.lambdify(x, f, 'numpy')
df_num = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0, 100, 100)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_num(x_vals), label=f'f(x) = {f}', color='blue')
plt.plot(x_vals, df_num(x_vals), label=f"f'(x) = {df}", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Тобто бачимо, що похідна $ f'(x) = 2x - 5 $ показує лінійну швидкість зростання витрат, а сама, при менших $ x $ (кількості вироблених одиниць) приріст витрат менший, а при більших $ x $ кожна додаткова одиниця виробництва коштує дедалі дорожче.
