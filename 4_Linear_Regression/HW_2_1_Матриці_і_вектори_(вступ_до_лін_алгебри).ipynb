
### Завдання 3: Множення матриць
1. Задані дві матриці `A` і `B`розмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю `result_matrix` на екран.

2. Заданий вектор `C` розмірністю 2х1. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на вектор `C` і який результат буде. А потім виконайте множення і виведіть результат на екран.

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
# Щоб результуюча матриця була розміром 2x2 потрібно B помножити на A
result_matrix = np.dot(B, A)
print('Множення матриць BA:')
display(result_matrix, result_matrix.shape)
```

```python
# В даному випадку можемо помножити result_matrix з розмірністю 2х2 на матрицю C розмірністю 2x1 (навпаки не вийде)
# Розмірність результуючої матриці буде 2x1
result_matrix_C = np.dot(result_matrix, C)
print('Множення result_matrix на C:')
display(result_matrix_C, result_matrix_C.shape)
```

```python
# В даному випадку можемо помножити result_matrix з розмірністю 2х2 на матрицю D розмірністю 2x2 (можна і навпаки, але результат буде інший)
# Розмірність результуючої матриці буде 2x2
result_matrix_D = np.dot(result_matrix, D)
print('Множення result_matrix на D:')
display(result_matrix_D, result_matrix_D.shape)
```

```python
print('Матриця result_matrix:')
display(result_matrix)
```

```python
result_matrix_T = result_matrix.T
print('Транспонована матриця до result_matrix:')
display(result_matrix_T)
```

```python
vector_rand_T = vector_rand.T
print('Початковий вектор:')
display(vector_rand, vector_rand.shape)
print('Транспонований вектор:')
display(vector_rand_T, vector_rand_T.shape)
# Бачимо, що змін не відбулося, адже для NumPy цей вектор - це 1-вимірний масив, а не стовпчик або рядок
```

```python
column_vector = np.expand_dims(vector_rand, axis=1)
print('Використання expand_dims:')
display(column_vector)
display(vector_rand.shape, column_vector.shape)
# Бачимо, що відбулося перетворення одновимірного вектора у стовпчиковий вектор з розміром (10, 1)
```

```python
column_vector_T = column_vector.T
print('Транспонований вектор column_vector:')
display(column_vector_T,column_vector_T.shape)
# Отже, транспонування працює коректно для двох і вище вимірних масивів
```
